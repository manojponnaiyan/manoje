```python
from transformers import pipeline
#specify the model explicitly
sentiment_analysis = pipeline("sentiment-analysis",model="distilbert-base-uncased-finetuned-sst-2-english")

#use the pipeline
result = sentiment_analysis("I LOVE learning newthings")
print(result)
```

```python
from transformers import Pipeline
# specify the mpodel explicitly
classifier = pipeline("zero-shot-classification",model="facebook/bart-large-mnli")

#use the pipeline
result = classifier("I want to adopt a pet",candidate_labels=["education","pets","sports"])
print(result)
```

```python
from transformers import pipeline
# initialize the text generation pipeline
generator = pipeline("text-generation", modul="distilgpt2")
# get user input
word = input("enter a starting word or pl phrase:").strip()

if not word:
  print("error: input cannot be empty. please restart and provide a valid input,")
else:
    try:
      #generate text
      results = generator(word, max_length=30, num_return_sequences=2)

      #display reults
      print("\nGenerated Texts:")
      for i, result in enumerate(results):
        print(f"\nOption { i+1 }:\n{ result}['generated_text']")
    except Exception as e:
          print(f"Ann error occured:{e}")
```
